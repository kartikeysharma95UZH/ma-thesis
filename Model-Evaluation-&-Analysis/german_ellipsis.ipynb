```python
# /home/user/ksharma/ks_thesis/ma-thesis/result_evaluation/data_files/english_ellipsis
# pd.read_csv("/home/user/ksharma/ks_thesis/saved_data_files/updated_self_model/flan-t5-large_ep-10_ga-1_b-4_lr-0.0001_ENGLISH_ELLIPSIS_20230725210509.csv")

import os
import glob
import pandas as pd

# Directory path containing the CSV files
directory_path_german = "/home/user/ksharma/ks_thesis/ma-thesis/result_evaluation/data_files/german_ellipsis"

# Use glob to get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(directory_path_german, "*.csv"))
print(csv_files)
# Initialize lists to store results
file_names = []
near_match_list = []
exact_match_list = []
avg_levenshtein_list = []

# Loop through each CSV file and calculate metrics
for csv_file in csv_files:
    if csv_file == '/home/user/ksharma/ks_thesis/ma-thesis/result_evaluation/data_files/german_ellipsis/german_results.csv':
        continue
    # Read the CSV file
    df_german = pd.read_csv(csv_file)
    # df = df[:100]  # Limit the number of rows for faster processing
    # print(df_german.columns)
    # Calculate metrics
    near_match = len(df_german[df_german['Levenshtein_dist'] < 3]) / len(df_german)
    exact_match = len(df_german[df_german['Levenshtein_dist'] == 0]) / len(df_german)
    avg_levenshtein = df_german['Levenshtein_dist'].mean()
    
    # Extract relevant parts from the filename
    filename = os.path.basename(csv_file)
    model_with_lr = filename.split("_ELLIPSIS")[0]
    
    # Append results to the lists
    file_names.append(model_with_lr)
    near_match_list.append(near_match)
    exact_match_list.append(exact_match)
    avg_levenshtein_list.append(avg_levenshtein)

# Create a DataFrame with the results
results_df_german = pd.DataFrame({
    'Model with LR': file_names,
    'Near Match': near_match_list,
    'Exact Match': exact_match_list,
    'Average Levenshtein': avg_levenshtein_list
})

results_df_sorted_german = results_df_german.sort_values(by='Exact Match', ascending=False)
# results_df_filtered = results_df_sorted[results_df_sorted['Model with LR'] != "mt5-base_ep-10_ga-1_b-4_lr-0.0001_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "flan-t5-large_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "t5-large_ep-10_ga-1_b-4_lr-0.0001_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "t5-base_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "flan-t5-base_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
results_df_sorted_german
```

## ENGLISH ELLIPSIS

```python
# /home/user/ksharma/ks_thesis/ma-thesis/result_evaluation/data_files/english_ellipsis
# pd.read_csv("/home/user/ksharma/ks_thesis/saved_data_files/updated_self_model/flan-t5-large_ep-10_ga-1_b-4_lr-0.0001_ENGLISH_ELLIPSIS_20230725210509.csv")

import os
import glob
import pandas as pd

# Directory path containing the CSV files
directory_path_english = "/home/user/ksharma/ks_thesis/ma-thesis/result_evaluation/data_files/english_ellipsis"

# Use glob to get a list of all CSV files in the directory
csv_files = glob.glob(os.path.join(directory_path_english, "*.csv"))
print(csv_files)
# Initialize lists to store results
file_names = []
near_match_list = []
exact_match_list = []
avg_levenshtein_list = []

# Loop through each CSV file and calculate metrics
for csv_file in csv_files:
    # Read the CSV file
    df = pd.read_csv(csv_file)
    df = df.drop(df[(df['Unnamed: 0'] == 488) | (df['Unnamed: 0'] == 499)].index)
    # df = df[:100]  # Limit the number of rows for faster processing
    
    # Calculate metrics
    near_match = len(df[df['Levenshtein_dist'] < 3]) / len(df)
    exact_match = len(df[df['Levenshtein_dist'] == 0]) / len(df)
    avg_levenshtein = df['Levenshtein_dist'].mean()
    
    # Extract relevant parts from the filename
    filename = os.path.basename(csv_file)
    model_with_lr = filename.split("_ENGLISH_ELLIPSIS")[0]
    
    # Append results to the lists
    file_names.append(model_with_lr)
    near_match_list.append(near_match)
    exact_match_list.append(exact_match)
    avg_levenshtein_list.append(avg_levenshtein)

# Create a DataFrame with the results
results_df_english = pd.DataFrame({
    'Model with LR': file_names,
    'Near Match': near_match_list,
    'Exact Match': exact_match_list,
    'Average Levenshtein': avg_levenshtein_list
})

results_df_sorted_english = results_df_english.sort_values(by='Exact Match', ascending=False)
# results_df_filtered = results_df_sorted[results_df_sorted['Model with LR'] != "mt5-base_ep-10_ga-1_b-4_lr-0.0001_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "flan-t5-large_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "t5-large_ep-10_ga-1_b-4_lr-0.0001_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "t5-base_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
# results_df_filtered = results_df_filtered[results_df_filtered['Model with LR'] != "flan-t5-base_ep-10_ga-1_b-4_lr-0.0003_wd-0.01"]
results_df_sorted_english
```
