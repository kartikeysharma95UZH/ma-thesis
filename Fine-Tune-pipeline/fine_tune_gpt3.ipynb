# Functions :

### Data Preparation

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def data_preparation(dataset_name):
    data = pd.read_csv(dataset_name)
    valid_rows = data[data['valids'] == 1.0]
    valid_data_df = valid_rows.copy()
    fine_tuning_data = valid_data_df[['ellipsis', 'corrected']]
    return fine_tuning_data
```

### Test/Train Split

```python
def split_test_train(dataset):
    # train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)
    """ Change this later """
    fine_tuning_dataset = dataset.copy()
    train_data = fine_tuning_dataset[:400]
    # test_data = fine_tuning_dataset.tail(100)
    test_data = fine_tuning_dataset.loc[415:520]
    """ Change this later """
    
    return train_data, test_data
```

### Preprocess

```python
def preprocess_finetune(train_data, test_data):

    train_data = train_data.rename(columns={"corrected" : "completion"})
    train_data = train_data.rename(columns={"ellipsis" : "prompt"})

    test_data = test_data.rename(columns={"corrected" : "completion"})
    test_data = test_data.rename(columns={"ellipsis" : "prompt"})

    train_data['prompt'] = train_data['prompt'].apply(lambda x: " " + x + " ->" if not pd.isnull(x) else x)
    train_data['completion'] = train_data['completion'].apply(lambda x: " " + x + " ###" if not pd.isnull(x) else x)

    test_data['prompt'] = test_data['prompt'].apply(lambda x: " " + x + " ->" if not pd.isnull(x) else x)
    # test_data['completion'] = test_data['completion'].apply(lambda x: " " + x if not pd.isnull(x) else x)
    

    return train_data, test_data
```

### Prediction 

```python
def predict(test_data, model, temperature):
    import openai
    import json
    generated_text = []
    openai.api_key = "sk-xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
    for sample in test_data["prompt"]:
        # print(sample["prompt"])
        # print(sample)
        # Call the OpenAI API to generate text using your fine-tuned GPT-3 model
        response = openai.Completion.create(
            engine = model, 
            prompt = sample,
            max_tokens = 300,
            temperature = temperature,
            # n= "n",
            stop = " ###",
        )
        # Append the generated text to the list
        generated_text.append(response.choices[0].text)

    generated_text_without_spaces = [w.lstrip() for w in generated_text]
    return generated_text_without_spaces
```

### Classification:

```python
def accuracy_metric(df):
    y_true = df['expected_text'].tolist()
    y_pred = df['generated_text'].tolist()
    correct = 0
    total = len(y_true)
    for true, predict in zip (y_true, y_pred):
        if true == predict:
            correct += 1

    accuracy = float(correct/total)

    return accuracy
```

## Final Evaluation of "ada"

```python
# dataset_name = "german-binomial-hyphenated-compounds - Sheet1.csv"
# model_name = "ada:ft-personal-2023-02-13-17-15-27"
# model_temp = 0.6

def get_accuracy(dataset_name, model_name, model_temp):
    df = data_preparation(dataset_name)
    train_data, test_data = split_test_train(df)
    train_data, test_data = preprocess_finetune(train_data, test_data)
    subset_test_df = test_data.sample(n=30, random_state=52)
    generated_text = predict(subset_test_df, model_name, model_temp)
    accuracy_df = subset_test_df.copy()
    accuracy_df['generated_text'] = generated_text
    accuracy_df = accuracy_df.rename(columns={"completion" : "expected_text"})
    accuracy = accuracy_metric(accuracy_df)
    return accuracy, accuracy_df
```

```python
dataset_name = "german-binomial-hyphenated-compounds - Sheet1.csv"
model_name = "ada:ft-personal-2023-02-13-17-15-27"
model_temp = 0

accuracy_ada, accuracy_ada_df = get_accuracy(dataset_name, model_name, model_temp)
```
